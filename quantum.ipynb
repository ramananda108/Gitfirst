```python
import numpy as np
import matplotlib.pyplot as plt

h = 6.626e-34  # Planck's constant (J.s)
m = 9.11e-31   # Electron mass (kg)
L = 1e-9       # Length of box = 1 nm

n = np.arange(1, 6)  # Quantum numbers 1 to 5
E = (n**2 * h**2) / (8 * m * L**2)

plt.stem(n, E, basefmt=" ")
plt.title("Energy Levels of Particle in 1D Box")
plt.xlabel("Quantum Number n")
plt.ylabel("Energy (Joules)")
plt.grid(True)
plt.show()
```
