```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-20, 20, 1000)
psi = np.exp(1j * x)  # free particle wavefunction
plt.plot(x, psi.real, label='Real')
plt.plot(x, psi.imag, label='Imaginary')
plt.title("Free Particle Wavefunction")
plt.legend()
plt.show()
```

2. Probability Density for a Particle in a Box (1D)

```python
import numpy as np
import matplotlib.pyplot as plt

L = 1
n = 1
x = np.linspace(0, L, 1000)
psi_n = np.sqrt(2/L) * np.sin(n * np.pi * x / L)
plt.plot(x, psi_n**2)
plt.title("Probability Density: Particle in 1D Box (n=1)")
plt.xlabel("x")
plt.ylabel("|psi|^2")
plt.show()
```

3. Energy Levels of the Hydrogen Atom

```python

Z = 1  # atomic number
n = np.arange(1, 6)
E_n = -13.6 * Z**2 / n**2
for i, e in enumerate(E_n):
    print(f"n={n[i]}: E={e:.2f} eV")
```

4. Harmonic Oscillator Eigenstates

```python
from scipy.special import hermite
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-4, 4, 500)
n = 2
psi_n = (1/np.sqrt(2**n * math.factorial(n) * np.sqrt(np.pi))) * hermite(n)(x) * np.exp(-x**2/2)
plt.plot(x, psi_n)
plt.title(f"Harmonic Oscillator Eigenstate (n={n})")
plt.xlabel("x")
plt.ylabel("psi_n(x)")
plt.show()
```

5. Superposition of Two States

```python

import numpy as np
import matplotlib.pyplot as plt

L = 1
x = np.linspace(0, L, 1000)
psi1 = np.sqrt(2/L) * np.sin(np.pi * x / L)
psi2 = np.sqrt(2/L) * np.sin(2 * np.pi * x / L)
psi = (psi1 + psi2) / np.sqrt(2)
plt.plot(x, psi**2)
plt.title("Probability Density: Superposition (n=1 & n=2)")
plt.xlabel("x")
plt.ylabel("|psi|^2")
plt.show()
```

6. Heisenberg Uncertainty Principle Visualization

```python

import numpy as np
import matplotlib.pyplot as plt

sigma_x = np.linspace(0.1, 5, 100)
hbar = 1
sigma_p = hbar / (2 * sigma_x)
plt.plot(sigma_x, sigma_p)
plt.title("Heisenberg Uncertainty Principle")
plt.xlabel("Δx")
plt.ylabel("Δp")
plt.show()
```

7. Time Evolution of a Gaussian Wave Packet

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
t = 0.5
sigma = 1
hbar = 1
m = 1
x0 = 0
k0 = 5

psi = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-x0)**2/(2*sigma**2)) * np.exp(1j*k0*x)
psi_t = psi * np.exp(-1j * hbar * k0**2 * t / (2*m))
plt.plot(x, np.abs(psi_t)**2)
plt.title("Gaussian Wave Packet (Time Evolution)")
plt.xlabel("x")
plt.ylabel("|psi(x,t)|^2")
plt.show()
```

8. Measurement Probabilities in Spin-1/2 Systems

```python

import numpy as np

from numpy import array, dot, conj
sigma_z = array([[1, 0], [0, -1]])
up = array([[1], [0]])
down = array([[0], [1]])

pz_up = np.abs(dot(conj(up.T), dot(sigma_z, up)))[0,0]
pz_down = np.abs(dot(conj(down.T), dot(sigma_z, down)))[0,0]
print("Probabilities for σ_z measurement:")
print("Up:", pz_up)
print("Down:", pz_down)
```
