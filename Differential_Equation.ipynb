1 Solution of First Order Differential Equation by ODEINT

```python
# 1. Solution of First Order Differential Equation by ODEINT
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math

def f(y, x):
    return math.exp(-x)

y1 = 0
xs = np.arange(0, 10, 0.05)
ys = odeint(f, y1, xs)

plt.plot(xs, ys, '.b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of 1st ODE')
plt.show()
```

2 Solution of First Order ODE: dy/dx = x² - e**2

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math

def f(y, x):
    return x**2 - math.exp(-x)

y1 = 0
xs = np.arange(0, 10, 0.05)
ys = odeint(f, y1, xs)

plt.plot(xs, ys, '.b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of 1st Order ODE')
plt.show()
```

Second Order Differential Equation: d²y/dx² + 2 dy/dx = −y

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u, x):
    return (u[1], -2*u[1] - u[0])

y0 = [0, 1]
xs = np.linspace(0, 10, 1000)
us = odeint(f, y0, xs)
ys, dys = us[:, 0], us[:, 1]

plt.plot(xs, ys, label='Function y')
plt.plot(xs, dys, label='Derivative dy/dx')
plt.xlabel('x')
plt.ylabel('y, dy/dx')
plt.title('Second Order ODE Solution')
plt.legend()
plt.show()
```

4 Second Order Differential Equation with exponential term:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u, x):
    return (u[1], -np.exp(-x)*u[1] - u[0])

y0 = [0, 1]
xs = np.linspace(0, 10, 1000)
us = odeint(f, y0, xs)
ys, dys = us[:, 0], us[:, 1]

plt.plot(xs, ys, label='Function y')
plt.plot(xs, dys, label='Derivative dy/dx')
plt.xlabel('x')
plt.ylabel('y, dy/dx')
plt.title('Second Order ODE (exp term)')
plt.legend()
plt.show()
```

5 Dirac Delta Integration

```python
import numpy as np
from scipy.integrate import simpson

g = lambda x: x**2 + 4
mu, sigma = 2, 0.001
f = lambda x: 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2 / (2*sigma**2))
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 1000)
I = simpson(f(x)*g(x), x)
print("Value of the Integration is:", I)
```
